```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
page=0
data={'product name':[],'image url 1':[],'image url 2':[],'price':[],'category':[]}
links=[]
```

```python
while(True):
    page+=1
    response = requests.get(f'https://www.khaadi.com/pk/woman.html?p={page}')
    soup = BeautifulSoup(response.text,'html.parser')
    for li in soup.find_all('li',{'class':'item product product-item'}):
        data['product name'].append(li.find('a',{'class':'product-item-link'}).text.strip())
        src_href = li.find('a',{'class':'product photo product-item-photo'})
        data['image url 1'].append(src_href.find_all('img')[0].get('src'))
        data['image url 2'].append(src_href.find_all('img')[1].get('src'))
        links.append(src_href.get('href'))
        data['price'].append(li.find('span',{'class':'price'}).text)
    print(str(page) +' Done')
    if not soup.find('a',{'class':'action next'}):
        for p,link in enumerate(links,start=1): # For Category
            response2 = requests.get(link)
            soup2 = BeautifulSoup(response2.text,'html.parser')
            category = soup2.find('div',{'class':'value','itemprop':"description"})
            data['category'].append(category.text) if category else data['category'].append(None)
            print(str(len(links)-p) +' Category Remaining')
        break
```

```python
data=pd.DataFrame.from_dict(data)
```

```python
data=pd.DataFrame.from_dict(data)
data.to_csv('Khaadi ')
```
