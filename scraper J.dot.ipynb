```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
page=0
links=[]
data={'product name':[],'image url':[],'product category':[],'price':[]}
```

```python
while(True):
    page += 1
    response =  requests.get(f'https://www.junaidjamshed.com/mens.html?p={page}')
    soup = BeautifulSoup(response.text,'html.parser')
    for li in soup.find_all('li', {'class':'item product product-item'}):
        data['product name'].append(li.find('a', {'class':'product-item-link'}).text.strip())
        data['image url'].append(li.find('img',{'class':'photo image'}).get('src'))
        data['price'].append(li.find('span',{'class':'price'}).text)
        #links.append(li.find('a',onclick=True,href=True).get('href'))
        soup2 = BeautifulSoup(requests.get(li.find('a',onclick=True,href=True).get('href')).text,'html.parser')
        category = soup2.find('td',{'class':'col data','data-th':'Product Category'})
        if category: data['product category'].append(category.text)
        else:
            category = soup2.find('div',{'class':'value','itemprop':'description'})
            data['product category'].append(category.text) if category else data['product category'].append(None)
    print(str(page)+' Done')
    if not soup.find_all('a',{'class': 'action next'}):
        break
data = pd.DataFrame.from_dict(data)
```

```python
data.dropna(axis=0,how='all')
```
